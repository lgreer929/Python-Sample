```python
# import orders and products data set
df_ords_prods_merge =  pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_new_columns.pkl'))
```

```python
# limiting dataframe to first 1,000,000 rows
df = df_ords_prods_merge[:1000000]
```

```python
# rename product_name column
df = df.rename(columns={'product_name': 'prod_name'})
```

```python
# group dataframe by prod_name
grouped = df.groupby('prod_name')
```

```python
grouped.head()
```

```python
# show column names
print(df.columns.tolist())
```

```python
print(df['product_name'].dtype)
```

```python
# group department and average number of orders
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# group department and average number of orders
df.groupby('department_id')['order_number'].mean()
```

```python
# group department and average, minimum, and maximum number of orders
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# create a new max_order column for most orders by specified customer
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head(15)
```

```python
# create criteria for customer loyalty flags
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# find value counts of customer loyalty categories
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# sort dataframe by user_id then loyalty_flag then order_number
df_ords_prods_merge_sorted = df_ords_prods_merge.sort_values(
    by=['user_id', 'loyalty_flag', 'order_number'],
    ascending=[True, False, False]
)
```
