# Data Aggregation

```python
# import libraries
import pandas as pd
import numpy as np
import os
```

```python
# create shortcut
path = r'C:\Users\lisag\Desktop\CF Work\Immersion 4 Tasks\Instacart (IC) Basket Analysis'
```

```python
# import pickle file orders and products
df_opm =  pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_new_columns.pkl'))
```

```python
# check output
df_opm.shape
```

```python
# find the mean for order_number as grouped by department_id
df_opm.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# create a new column of max order by which to sort customer loyalty
df_opm['max_order'] = df_opm.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_opm.head(25)
```

```python
# create criteria for customer rankings
# loyal customers are considered those who have order 40 + times
df_opm.loc[df_opm['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# regular customers are considered those who have ordered 11-40 times
df_opm.loc[(df_opm['max_order'] <= 40) & (df_opm['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# new customers are considered those who have ordered 10 or fewer times
df_opm.loc[df_opm['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# number of customers based on loyalty status
df_opm['loyalty_flag'].value_counts(dropna = False)
```

```python
# general statistics for loyalty_flag column
df_opm['loyalty_flag'].describe().round(1)
```

```python
# summary statistics for prices in the loyal customer category
df_opm[df_opm['loyalty_flag'] == 'Loyal customer']['prices'].describe().round(1)
```

```python
# summary statistics for prices in the regular customer category
df_opm[df_opm['loyalty_flag'] == 'Regular customer']['prices'].describe().round(1)
```

```python
# summary statistics for prices in the new customer category
df_opm[df_opm['loyalty_flag'] == 'New customer']['prices'].describe().round(1)
```

```python
# create a new column of order cost to sort high and low spenders
df_opm['order_cost'] = df_opm['order_number'] * df_opm['prices']
```

```python
df_opm.head()
```

```python
# calculating the average cost per user_id
df_opm['avg_order_cost'] = df_opm.groupby('user_id')['order_cost'].transform('mean').round(1)
```

```python
df_opm.head()
```

```python
# low spenders are considered those who have average orders less than $10
df_opm.loc[df_opm['avg_order_cost'] < 10, 'spender_type'] = 'Low spender'
```

```python
# high spenders are considered those who have average orders $10 or more
df_opm.loc[df_opm['avg_order_cost'] >= 10, 'spender_type'] = 'High spender'
```

```python
# number of customers based on spending type
df_opm['spender_type'].value_counts(dropna = False)
```

```python
# create a new column to determine median orders of users
df_opm['median_days_between_orders'] = df_opm.groupby('user_id')['days_since_prior_order'].transform('median').round(1)
```

```python
df_opm.head()
```

```python
# non-frequent customers are considered those whose orders are more than 20 days apart
df_opm.loc[df_opm['median_days_between_orders'] > 20, 'customer_frequency'] = 'Non-frequent customer'
```

```python
# regular customers are considered those whose orders are between 11 and 20 days apart
df_opm.loc[(df_opm['median_days_between_orders'] > 10) & (df_opm['median_days_between_orders'] <=20), 'customer_frequency'] = 'Regular customer'
```

```python
# frequent customers are considered those whose orders are less than 10 days apart
df_opm.loc[df_opm['median_days_between_orders'] < 10, 'customer_frequency'] = 'Frequent customer'
```

```python
# number of customers based on frequency of orders
df_opm['customer_frequency'].value_counts(dropna = False)
```

```python
# export as pickle file
df_opm.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_flagged.pkl'))
```
