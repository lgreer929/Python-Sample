# Cleaning Final Data

```python
# importing libraries
import pandas as pd
import numpy as np
import os
```

```python
# create path shortcut
path = r'C:\Users\lisag\Desktop\CF Work\Immersion 4 Tasks\Instacart (IC) Basket Analysis\02 Data\Prepared Data'
```

```python
df = pd.read_pickle(os.path.join(path, 'orders_products_flagged.pkl'))
```

```python
# quality check
df.head()
```

```python
# quality check
df.shape
```

```python
# finding missing values
df.isnull().sum()
```

```python
# identifying null values
df_nan_dspo = df[df['days_since_prior_order'].isnull() == True]
print(df_nan_dspo.sample(20))
```

```python
# replacing NaN values in days_since_prior_order with 0 to identify first time customers
df['days_since_prior_order'] = df['days_since_prior_order'].fillna(0)
```

```python
# checking results
df['days_since_prior_order'].isna().sum()
```

```python
# identifying more null values
df_nan_dspo_2 = df[df['median_days_between_orders'].isnull() == True]
print(df_nan_dspo_2.sample(5))
```

```python
# eliminating rows with null values in median_days_between_orders column
df = df.dropna(subset=['median_days_between_orders'])
```

```python
# checking results
df['median_days_between_orders'].isna().sum()
```

```python
# identifying more null values
df_nan_dspo_3 = df[df['customer_frequency'].isnull() == True]
print(df_nan_dspo_3.sample(20))
```

```python
# fixing logic mistake from previous df
# non-frequent customers are considered those whose orders are more than 20 days apart
df.loc[df['median_days_between_orders'] > 20, 'customer_frequency'] = 'Non-frequent customer'
```

```python
# regular customers are considered those whose orders are between 11 and 20 days apart
df.loc[(df['median_days_between_orders'] > 10) & (df['median_days_between_orders'] <=20), 'customer_frequency'] = 'Regular customer'
```

```python
# frequent customers are considered those whose orders are less than 10 days apart
df.loc[df['median_days_between_orders'] <= 10, 'customer_frequency'] = 'Frequent customer'
```

```python
# checking results
df['customer_frequency'].isna().sum()
```

```python
# export as pickle file
df.to_pickle(os.path.join(path, 'final_merged_cleaned.pkl'))
```
